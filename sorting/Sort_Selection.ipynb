```python
#Selection sort
#Time Complexity:  O(n2).
#Auxiliary Space: O(1)


def selectionSort(array, size):

    for step in range(size):
        min_idx = step
        
        for i in range(step + 1, size):
            if array[i] < array[min_idx]:
                min_idx = i

        (array[step], array[min_idx]) = (array[min_idx], array[step])

data = [ -2, 45, 0, 11, -9]
size = len(data)
selectionSort(data, size)
print('Sorted in Ascending order:')
print(data)
```

```python
Array = [ -2, 45, 0, 11, -9]

for i in range(len(Array)):   #loop to Traversee through all the elements in the given array
    min_idx = i   #setting min_idx equal to the first unsorted element
    
    for j in range(i + 1, len(Array)):   #loop to iterate over unsorted subarray
        if Array[min_idx] > Array[j]:    #Finding the minimum element in the unsorted sub-array
            min_idx = j

    Array[i], Array[min_idx] = Array[min_idx], Array[i]  #swaping the minimum element with the element ata min_idx

print(Array)
      
```
