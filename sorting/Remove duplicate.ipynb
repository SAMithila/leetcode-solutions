```python
def removeDuplicates(arr, n):
    if n == 0 or n == 1:
        return n

    temp = list(range(n))

    j = 0
    for i in range(0, n-1):
        if arr[i] != arr[i+1]:
            temp[j] = arr[i]
            j += 1

    temp[j] = arr[n-1]
    j += 1

    for i in range(0, j):
        arr[i] = temp[i]

    return j


```

```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
          s = set(nums)
          nums.clear()
          for i in s:
               nums.append(i)
          nums.sort()
          return len(nums)


# Create a list with duplicates
nums = [3, 1, 2, 3, 1, 4, 5, 2]

# Create an instance of Solution
sol = Solution()

# Call removeDuplicates and get the result
new_length = sol.removeDuplicates(nums)

# Print the result
print("Length after removing duplicates:", new_length)
print("List after removing duplicates:", nums)
```
