```python
#Brute force, Sorting
#Time Complexity: O(nlogn), where n is the size of input array.
#Auxiliary space: O(1), as no extra space is required.

def print2large(arr, arr_size):
    arr.sort(reverse = True)    #Sort the array in desending order
   
    for i in range (1, arr_size):    #Start from the second last element as the first element is the largest
        if (arr[i] != arr[0]):       #If the element is not equal to largest element
            print("This second largest element is", arr[i])
            return
    print("There is no second largest element")

arr = [12, 35, 1, 10, 34, 1]
n = len(arr)
print2large(arr, n)
            
```

```python
#Better approach

def find_second_largest(a, n):
    first_largest = -9999999      #A very small initial value
    for i in range(n):
        if first_largest < a[i]: 
            first_largest = a[i]    #Update first_largest if current element is larger

    
    second_largest = -999999     #A very small initial value
    for i in range(n):
        if a[i] > second_largest and a[i] < first_largest: # Update second_largest if conditions are met
            second_largest = a[i]

    return second_largest

if __name__ == '__main__':
    a = [12, 75, 1, 10, 74, 1]
    n = len(a)
    answer = find_second_largest(a, n)
    print("Second element", answer)
```
