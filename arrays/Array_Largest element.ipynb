```python
# Optimal solution / Native approach
# Time complexity O(n)
# Space complexity O(1)

def largest(arr, n):
    max = arr[0]    #Initialize maximum element

    for i in range(1, n):     #Traverse array elements from second and compare every element with current max
        if arr[i] > max:
            max = arr[i]
    return max

arr = [10, 324, 45, 90, 9808]
n = len(arr)
Ans = largest(arr, n)
print("Largest in given array ", Ans)

```

```python
# Brute Force solution / Sort function
# Time complexity O(n log n) 
# Space complexity O(1)

def largest(arr, n):
    arr.sort()     #Sort the array
    return arr[n-1]   # Return the last element, which is the largest

arr = [10, 324, 45, 90, 9808]
n = len(arr)
Ans = largest(arr, n)
print("Largest in given array ", Ans)
```

```python
# Optimal solution / reduce() function
# Time complexity O(n)
# Space complexity O(1)

from functools import reduce

def largest(arr):
    ans = reduce(max, arr)   # Sort the array
    return ans      # or returning largest value

arr = [10, 324, 45, 90, 9808]
n = len(arr)
Ans = largest(arr)
print("Largest in given array ", Ans)
```
