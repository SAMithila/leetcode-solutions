```python
def remove_duplicates(arr):
    n = len(arr)
    if n <= 1:
        return n

    unique_elements = set(arr)       #Use a set to store unique elements

    arr[:n] = unique_elements        #Update the originals array with unique elements

    return len(unique_elements)       #Return the count of unique elements


#Driver code
if __name__ == "__main__":
    arr = [1, 2, 3, 3, 4, 4, 4, 5, 5]

    n = remove_duplicates(arr)       #Remove duplicates and get the count of unique elements


    for i in range(n):               #Print the modified array containing unique elements 
        print(arr[i], end=" ")
```

```python
#Better approach

def remove_duplicates(arr):
    seen = []
    for elem in arr:
        if elem not in seen:
            seen.append(elem)
    arr[:len(seen)] = seen
    return len(seen)

#Driver code
if __name__ == "__main__":
    arr = [1, 2, 3, 3, 4, 4, 4, 5, 5]

    n = remove_duplicates(arr)       #Remove duplicates and get the count of unique elements


    for i in range(n):               #Print the modified array containing unique elements 
        print(arr[i], end=" ")
```

```python
def removeDuplicates(items):
    list1 = []
    for i in items:
        if i not in list1:
            list1.append(i)
    return list1

nums = [0, 0, 1, 1, 1, 2, 2, 3, 3, 4]
print(removeDuplicates(nums))
```
