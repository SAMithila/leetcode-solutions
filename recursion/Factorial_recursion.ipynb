```python
#Find the Factorial of a Number Using Recursive approach
#Find the Factorial of a Number Using One line Solution (Using Ternary operator): 

def factorial(n):
    return 1 if (n == 1 or n == 0) else n * factorial(n - 1)

num = 5
print("Factorial of", num, "is", factorial(num))
```

```python
#Find the Factorial of a Number Using Iterative approach_Example 1:

def factorial(n):
    if n < 0:
        return 0
    elif n == 0 or n == 1:
        return 1
    else:
        fact = 1
        while(n > 1):
            fact *=n
            n -= 1
        return fact

num = 5
print("Factorial of", num, "is", factorial(num))
```

```python
#Find the Factorial of a Number Using Iterative approach_Example 2:

def factorial(n):
    fact = 1
    for i in range(2, n + 1):
        fact *= i
    return fact

num = 5
print("Factorial of", num, "is", factorial(num))
```

```python
#Find the Factorial of a Number Using using In-built function 

import math

def factorial(n):
    return(math.factorial(n))

num = 5
print("Factorial of", num, "is", factorial(num))
```
