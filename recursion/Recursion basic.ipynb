```python
def func(i, n):
    if i > n:
        return
    print("Raj")
    func(i + 1, n)

def main():
    n = int(input("Enter a number: "))
    func(1, n)

main()
```

```python
#print 1 to n using recursion

def printNumber(n):
    if n > 0:    #check if n is greater than 0
        printNumber(n - 1)  #recursively call the printNumber function
        print(n, end=" ")   #print n

n = 3
printNumber(n)  #call the printNumber function
```

```python
#print n to 1 using recursion

def printReverseNumber(n):
    if n < 1:     #check if n is greater than 0
        return
    print(n, end=" ")   #print n
    printReverseNumber(n - 1)  #recursively call the printReverseNumber function
        

n = 3
printReverseNumber(n)  #call the printNumber function
```

```python
# Find sum of n natural numbers in Python with for loop

def calculate_sum(n):
    sum = 0

    for i in range(1, n + 1):
        sum += i

    return sum
    

num = int(input("Enter a number: "))

total_sum = calculate_sum(num)
print(f"The sum of the first {num} natural numbers is: {total_sum}")
```

```python
# Sum of first n natural numbers in Python with while loop

def calculate_sum(n):
    sum = 0
    i = 1

    while i <= n:
        sum += i
        i += 1

    return sum
    

num = int(input("Enter a number: "))

total_sum = calculate_sum(num)
print(f"The sum of the first {num} natural numbers is: {total_sum}")
```

```python
# Sum of n natural numbers in Python with recursion

def calculate_sum(n):
    if n == 1:
        return 1
    else:
        return n + calculate_sum(n - 1)

num = int(input("Enter a number: "))

total_sum = calculate_sum(num)
print(f"The sum of the first {num} natural numbers is: {total_sum}")
```
